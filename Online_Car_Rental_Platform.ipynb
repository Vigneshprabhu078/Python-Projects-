```python
import datetime

class CarRental:
    def __init__(self, stock=0):
        self.stock = stock

    def display_stock(self):
        print(f"Total cars available for rent: {self.stock}")
        return self.stock

    def rent_car(self, num_of_cars, basis):
        if num_of_cars <= 0:
            print("Number of cars should be positive!")
            return None
        elif num_of_cars > self.stock:
            print("Sorry, not enough cars available.")
            return None
        else:
            now = datetime.datetime.now()
            print(f"Rented {num_of_cars} car(s) on {basis} basis at {now}.")
            self.stock -= num_of_cars
            return now

    def return_car(self, rental_time, rental_basis, num_of_cars):
        if rental_time and rental_basis and num_of_cars:
            self.stock += num_of_cars
            now = datetime.datetime.now()
            rental_period = 0

            if rental_basis == 'hourly':
                rental_period = (now - rental_time).seconds / 3600
                bill = round(rental_period) * 5 * num_of_cars

            elif rental_basis == 'daily':
                rental_period = (now - rental_time).days
                bill = rental_period * 20 * num_of_cars

            elif rental_basis == 'weekly':
                rental_period = (now - rental_time).days / 7
                bill = rental_period * 60 * num_of_cars

            print(f"Thanks for returning your car. Your bill is ${bill:.2f}")
            return now
        else:
            print("You did not rent any car.")
            return None

class Customer:
    def __init__(self):
        self.cars = 0
        self.rental_basis = ''
        self.rental_time = None

    def request_car(self):
        cars = input("How many cars would you like to rent? ")
        try:
            cars = int(cars)
        except ValueError:
            print("Input should be a number.")
            return -1

        if cars < 1:
            print("Number of cars should be greater than zero.")
            return -1
        else:
            self.cars = cars
            return self.cars

    def return_car(self):
        if self.rental_time and self.rental_basis and self.cars:
            return self.rental_time, self.rental_basis, self.cars
        else:
            return None, None, None

def main():
    rental_system = CarRental(10)
    customer = Customer()

    while True:
        print("""
        ====== Car Rental Shop =======
        1. Display available cars
        2. Rent a car on hourly basis $5/hour
        3. Rent a car on daily basis $20/day
        4. Rent a car on weekly basis $60/week
        5. Return a car
        6. Exit
        """)

        choice = input("Enter your choice: ").strip()

        if choice == '1':
            rental_system.display_stock()

        elif choice == '2':
            customer.request_car()
            rental_time = rental_system.rent_car(customer.cars, 'hourly')
            if rental_time:
                customer.rental_basis = 'hourly'
                customer.rental_time = rental_time

        elif choice == '3':
            customer.request_car()
            rental_time = rental_system.rent_car(customer.cars, 'daily')
            if rental_time:
                customer.rental_basis = 'daily'
                customer.rental_time = rental_time

        elif choice == '4':
            customer.request_car()
            rental_time = rental_system.rent_car(customer.cars, 'weekly')
            if rental_time:
                customer.rental_basis = 'weekly'
                customer.rental_time = rental_time

        elif choice == '5':
            rental_time, rental_basis, num_of_cars = customer.return_car()
            rental_system.return_car(rental_time, rental_basis, num_of_cars)

        elif choice == '6':
            break

        else:
            print("Invalid input. Please enter a number between 1 and 6.")

if __name__ == "__main__":
    main()
```
