```python


# Sample data for users and admins
users_db = {'user1': 'password1'}
admins_db = {'admin1': 'adminpassword1'}

# Sample product catalog
product_catalog = [
    {'product_id': 1, 'name': 'Boots', 'category_id': 1, 'price': 100},
    {'product_id': 2, 'name': 'Coats', 'category_id': 1, 'price': 200},
    {'product_id': 3, 'name': 'Jackets', 'category_id': 1, 'price': 150},
    {'product_id': 4, 'name': 'Caps', 'category_id': 2, 'price': 50}
]

# Sample categories
categories = {1: 'Footwear', 2: 'Clothing'}

# Session storage
user_sessions = {}
admin_sessions = {}
cart = {}

def welcome_message():
    print("Welcome to the Simplilearn Shopping app")

def login(role):
    username = input(f"Enter {role} username: ")
    password = input(f"Enter {role} password: ")
    if role == 'user':
        if username in users_db and users_db[username] == password:
            session_id = f"user_{username}"
            user_sessions[session_id] = username
            print(f"User logged in with session ID: {session_id}")
            return session_id
        else:
            print("Invalid user credentials")
    elif role == 'admin':
        if username in admins_db and admins_db[username] == password:
            session_id = f"admin_{username}"
            admin_sessions[session_id] = username
            print(f"Admin logged in with session ID: {session_id}")
            return session_id
        else:
            print("Invalid admin credentials")
    return None

def view_catalog():
    for product in product_catalog:
        print(product)

def add_to_cart(session_id):
    if session_id not in user_sessions:
        print("Invalid user session")
        return
    product_id = int(input("Enter product ID to add: "))
    quantity = int(input("Enter quantity: "))
    if session_id not in cart:
        cart[session_id] = []
    cart[session_id].append({'product_id': product_id, 'quantity': quantity})
    print("Item added to cart")

def view_cart(session_id):
    if session_id not in user_sessions:
        print("Invalid user session")
        return
    if session_id in cart:
        print(cart[session_id])
    else:
        print("Cart is empty")

def remove_from_cart(session_id):
    if session_id not in user_sessions:
        print("Invalid user session")
        return
    product_id = int(input("Enter product ID to remove: "))
    if session_id in cart:
        cart[session_id] = [item for item in cart[session_id] if item['product_id'] != product_id]
        print("Item removed from cart")

def checkout(session_id):
    if session_id not in user_sessions:
        print("Invalid user session")
        return
    payment_option = input("Enter payment option (Net banking, PayPal, UPI): ")
    if session_id in cart:
        print(f"Checkout successful using {payment_option}. Your order is placed.")
        del cart[session_id]
    else:
        print("Cart is empty")

def admin_add_product(session_id):
    if session_id not in admin_sessions:
        print("Invalid admin session")
        return
    product_id = int(input("Enter new product ID: "))
    name = input("Enter product name: ")
    category_id = int(input("Enter category ID: "))
    price = float(input("Enter price: "))
    product = {'product_id': product_id, 'name': name, 'category_id': category_id, 'price': price}
    product_catalog.append(product)
    print("Product added to catalog")

def admin_update_product(session_id):
    if session_id not in admin_sessions:
        print("Invalid admin session")
        return
    product_id = int(input("Enter product ID to update: "))
    name = input("Enter new product name: ")
    category_id = int(input("Enter new category ID: "))
    price = float(input("Enter new price: "))
    updated_product = {'product_id': product_id, 'name': name, 'category_id': category_id, 'price': price}
    for i, product in enumerate(product_catalog):
        if product['product_id'] == product_id:
            product_catalog[i] = updated_product
            print("Product updated")
            return
    print("Product not found")

def admin_remove_product(session_id):
    if session_id not in admin_sessions:
        print("Invalid admin session")
        return
    product_id = int(input("Enter product ID to remove: "))
    global product_catalog
    product_catalog = [product for product in product_catalog if product['product_id'] != product_id]
    print("Product removed from catalog")

def admin_add_category(session_id):
    if session_id not in admin_sessions:
        print("Invalid admin session")
        return
    category_id = int(input("Enter new category ID: "))
    category_name = input("Enter new category name: ")
    categories[category_id] = category_name
    print("Category added")

def admin_remove_category(session_id):
    if session_id not in admin_sessions:
        print("Invalid admin session")
        return
    category_id = int(input("Enter category ID to remove: "))
    if category_id in categories:
        del categories[category_id]
        print("Category removed")
    else:
        print("Category not found")

# Example usage

welcome_message()

while True:
    role = input("Enter your role (user/admin): ").strip().lower()
    if role not in ['user', 'admin']:
        print("Invalid role. Please enter 'user' or 'admin'.")
        continue

    session_id = login(role)
    if session_id:
        while True:
            if role == 'user':
                action = input("Choose action: view_catalog, add_to_cart, view_cart, remove_from_cart, checkout, logout: ").strip().lower()
                if action == 'view_catalog':
                    view_catalog()
                elif action == 'add_to_cart':
                    add_to_cart(session_id)
                elif action == 'view_cart':
                    view_cart(session_id)
                elif action == 'remove_from_cart':
                    remove_from_cart(session_id)
                elif action == 'checkout':
                    checkout(session_id)
                elif action == 'logout':
                    del user_sessions[session_id]
                    print("Logged out.")
                    break
                else:
                    print("Invalid action.")
            elif role == 'admin':
                action = input("Choose action: view_catalog, add_product, update_product, remove_product, add_category, remove_category, logout: ").strip().lower()
                if action == 'view_catalog':
                    view_catalog()
                elif action == 'add_product':
                    admin_add_product(session_id)
                elif action == 'update_product':
                    admin_update_product(session_id)
                elif action == 'remove_product':
                    admin_remove_product(session_id)
                elif action == 'add_category':
                    admin_add_category(session_id)
                elif action == 'remove_category':
                    admin_remove_category(session_id)
                elif action == 'logout':
                    del admin_sessions[session_id]
                    print("Logged out.")
                    break
                else:
                    print("Invalid action.")
    else:
        print("Login failed.")
```
